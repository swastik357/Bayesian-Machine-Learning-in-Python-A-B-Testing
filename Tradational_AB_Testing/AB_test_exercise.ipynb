```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest
```

```python
df = pd.read_csv('advertisement_clicks.csv')
```

```python
df.head()
```

```python
df[df['advertisement_id'] == 'A'].head()
```

```python
x1 = df[df['advertisement_id'] == 'A']['action'].dropna().to_numpy()
```

```python
x2 = df[df['advertisement_id'] == 'B']['action'].dropna().to_numpy()
```

```python
sns.kdeplot(x1, label='A')
sns.kdeplot(x2, label='B')
plt.legend();
```

```python
x1.mean(), x2.mean()
```

```python
ztest(x1, x2, usevar='unequal')
```

```python
# MY COMMENTS: p-value = 0.00128<0.05, hence we reject NULL hypothesis (that respective means are same)
```
